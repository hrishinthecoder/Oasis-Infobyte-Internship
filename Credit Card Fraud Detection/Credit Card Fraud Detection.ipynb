Under Sampling: Building a sample dataset containing similar distribution of the normal tansctions and fradulent transactions.


```python
legit_sample = legit.sample(n=492)
```

```python
fraud_sample = fraud.sample(n=492)
```

```python
#concatenating two dataframes
new_df=pd.concat([legit_sample,fraud_sample],axis = 0) # axis 0 means rows, axis 1 means columns
```

```python
#Spliting the data into features and target
X = new_df.drop(columns = 'Class',axis = 1)
Y = new_df['Class']
```

```python
X
```

```python
Y.head()
```

```python
#spliting into training and testing data

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size = 0.2, stratify = Y, random_state = 2)

#0.2 means 20 percentage
#random_state will split the data in same way
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
#model training
#Using Logistic Regression for binry classification problems
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# accuracy on trining data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training Data:", training_data_accuracy )
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score on test data:", test_data_accuracy)
```
